```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
print("Successfully Imported")
```

### Read Data from .csv File

```python
# load the data
data = pd.read_csv("Spotify_data clean.csv")
data.head()
```

### Clean and Process Data:

```python
# data clean
data = data.dropna(subset=['artist_genres'])
data.head()
```

```python
# selecting only important columns
data = data[['track_name', 'track_popularity','artist_name','artist_popularity', 'artist_followers', 'album_name', 'album_release_date', 'track_duration_min']]
data.head()
```

```python
#Top 10 Artists Vs Number of their Songs Released
artists = data["artist_name"].value_counts().head(10)
artists
```

```python
x_bar = artists.values
y_bar = artists.index
plt.figure(figsize=(12,8))
plt.bar(y_bar, x_bar, color="purple")
plt.xticks(rotation=90, fontsize=10)
plt.ylim(0,350)
plt.xlabel("Artists", fontsize =15)
plt.ylabel("Total Number of Songs", fontsize =15)
plt.title("Top 10 Artists by Number of Songs")
plt.show()
```

```python
#Track distribution as per their popularity
songs_as_per_popularity = data["track_popularity"]
songs_as_per_popularity
```

```python
plt.figure(figsize=(10,8))
plt.hist(songs_as_per_popularity, bins=20, color="green", edgecolor="purple")
plt.title("Track Distribution Based on Popularity")
plt.ylabel("Number of Songs")
plt.xlabel("Popularity on the Scale (0-100) ")
plt.ylim(0,600)
plt.tight_layout()
plt.show()
```

```python
#Use Boxplot to determine the outliers(Tracks) based on Duration
plt.figure(figsize=(12,8))
sns.boxplot(data["track_duration_min"])
plt.title("Duration of Tracks")
plt.ylabel("Time Duratiion (Min)")
plt.show()
```

```python
# Relationship between Year and Number of Songs released
data["album_release_date"] = pd.to_datetime(data["album_release_date"])
data["album_release_year"] = data["album_release_date"].dt.year
year_count = data["album_release_year"].value_counts()
year_count
```

```python
plt.figure(figsize=(12,8))
plt.scatter(x=year_count.index, y=year_count.values)
plt.title("Number of Songs Released Per Year")
plt.ylim(0,450)
plt.xlabel("Release Year")
plt.ylabel("Number of Songs")
plt.tight_layout()
plt.show()
```
